```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load iris dataset
iris = sns.load_dataset('iris')
# display first 5 rows of the dataset
iris.head()
```

```python
# check shape of the dataset
print("Shape of the dataset:", iris.shape)
# check column names and data types
print("\nColumn names and data types:")
print(iris.info())
```

## Dataset Understanding & Description:
`Data Overview:` After loading the dataset, we observe the following characteristics:

* `Instances:` The dataset contains 150 rows (samples).
* `Attributes:` There are 4 numerical features (Sepal Length, Sepal Width, Petal Length, Petal Width).
* `Target Variable:` The species column represents the class of each flower.

##  Exploratory Data Analysis (EDA):

### Scatter Plot:

```python
# Scatter plot of Petal Length vs Petal Width
plt.figure(figsize=(10, 6))
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species', s=100)
plt.title('Relationship between Petal Length and Petal Width')
plt.show()
```

### Key Insights from Scatter Plot:
* `Setosa is Distinct:` The Setosa species forms a completely separate cluster. Both its petal length and width are significantly smaller than the other two species, making it the most easily identifiable variety.
* `Positive Correlation:` There is a strong positive linear relationship between petal_length and petal_width. As the length of the petal increases, the width tends to increase as well across all species.
* `Species Overlap:` While Setosa is isolated, there is noticeable overlap between Versicolor (orange) and Virginica (green), particularly where petal length ranges between 4.5 cm and 5.0 cm. This suggests that petal measurements alone might not be enough to perfectly distinguish these two species, and we may need to analyze sepal data for better clarity.

```python
# scatter plot of Sepal Length vs Sepal Width
plt.figure(figsize=(10, 6))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', s=100)
plt.title('Relationship between Sepal Length and Sepal Width')
plt.show()
```

## Insights: Sepal Length vs. Sepal Width:
* `Significant Overlap:` Unlike the petal measurements, the sepal data shows a high degree of overlap between the Versicolor and Virginica species. This suggests that these two species cannot be clearly distinguished based on sepal dimensions alone.
* `Setosa's Position:` Even though Setosa is still somewhat grouped together, it is not as isolated as it was in the petal plot. However, we can observe that Setosa generally has shorter but wider sepals compared to the other two species.
* `Low Predictability:` There is no strong linear correlation between sepal length and sepal width across the species, making these features less reliable for building a classification model compared to petal features.

### Histogram:

```python
# Histogram to examine data distribution.
# Histogram of Sepal Length
plt.figure(figsize=(10, 6))
sns.histplot(data=iris, x='sepal_length', hue='species', kde=True, bins=20)
plt.title('Distribution of Sepal Length by Species')
plt.show() 
```

### Insight: Distribution of Sepal Length:
* `Peak Density (Mode):` The histogram clearly shows that the most frequent sepal length for Setosa is around 5.0 cm, for Versicolor it is approximately 5.7 cm, and for Virginica, the peak is higher, around 6.4 cm (as indicated by the highest green bar).
* `Species Gradient:` We can observe a gradual shift in the distribution; as we move from Setosa to Virginica, the sepal length generally increases.
* `Overlap Analysis:` There is a significant overlap in the 5.5 cm to 6.5 cm range between all three species, which confirms that sepal length alone is not a perfect feature to distinguish between them, although it shows a clear trend.

```python
# histogram of Sepal Width
plt.figure(figsize=(10, 6))
sns.histplot(data=iris, x='sepal_width', hue='species', kde=True, bins=20)
plt.title('Distribution of Sepal Width by Species')
plt.show()
```

### Insight: Distribution of Sepal Width:
* `Distinct Peak for Setosa:` The histogram shows that Setosa has a higher average sepal width compared to the other species, with a clear mode (peak) around 3.4 cm (as shown by the highest blue bar).
* `Significant Overlap:` There is heavy overlapping between Versicolor and Virginica in the range of 2.5 cm to 3.2 cm. This makes it nearly impossible to distinguish these two species based only on sepal width.
* `Wider Sepals:` Interestingly, even though Setosa is the smallest flower in terms of petals, it tends to have the widest sepals among the three species.

```python
# histogram of petal length
plt.figure(figsize=(10, 6))
sns.histplot(data=iris, x='petal_length', hue='species', kde=True, bins=20)
plt.title('Distribution of Petal Length by Species')
plt.show()
```

### Insight: Distribution of Petal Length:
* `Distinct Peaks (Multimodality):` The histogram confirms that the dataset is multimodal. The Setosa species has a very sharp and isolated peak at approximately 1.5 cm, indicating very low variance in its petal length.

* `Species Gradient:` There is a clear progression in petal size across species: Setosa (1.5 cm) < Versicolor (~4.5 cm) < Virginica (~5.5 cm).

* `Minimal Overlap:` Unlike sepal measurements, there is almost zero overlap between Setosa and the other two species. While there is some overlap between Versicolor and Virginica, their peaks are sufficiently far apart to make petal length a primary feature for classification.


```python
# histogram of petal Width
plt.figure(figsize=(10, 6))
sns.histplot(data=iris, x='petal_width', hue='species', kde=True, bins=20)
plt.title('Distribution of Petal Width by Species')
plt.show()
```

### Insight: Distribution of Petal Width:
* `Extreme Isolation of Setosa:` The histogram shows that Setosa is completely separated from the other two species. Its petal width is very narrow, with a sharp peak (mode) at approximately 0.2 cm.
* `Clear Classification Boundary:` There is a significant gap between 0.6 cm and 1.0 cm where no data points exist. This "empty space" makes petal width an excellent feature for identifying Setosa with 100% accuracy.
* `Versicolor vs. Virginica:` Versicolor has a peak around 1.3 cm, while Virginica is wider, peaking around 1.9 cm to 2.0 cm. Although there is a small amount of overlap between these two, they are much more distinct here than they were in the sepal plots.
* `Consistency:` The very high bar for Setosa at 0.2 cm shows that this species is extremely consistent in its physical growth, whereas Virginica shows more variation (it's more spread out).

### Box Plot:

```python
# box plot to check for outliers or patterns in the data
plt.figure(figsize=(15, 10))

# Subplots for each feature
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
for i, feature in enumerate(features):
    plt.subplot(2, 2, i+1)
    sns.boxplot(x='species', y=feature, data=iris)
    plt.title(f'Box Plot of {feature}')

plt.tight_layout()
plt.show()
```

### Box Plot Insights:
* `Outlier Identification:` Outliers are primarily observed in the Sepal Width feature across all species, and in the Petal Length/Width for the Setosa species. However, these are minimal and likely represent natural biological variations rather than data entry errors.
* `Feature Discriminability:` The box plots for Petal Length and Petal Width show no overlap between Setosa and the other species, highlighting these as the most significant features for classification.
* `Data Variance:` Iris-Virginica shows the highest variance (largest box and whiskers) in almost all features, particularly in Petal Length, suggesting it has the most diverse physical characteristics.
* `Median Trends:` There is a clear upward trend in the median values from Setosa to Virginica for all features except Sepal Width, where Setosa actually has a higher median width.
